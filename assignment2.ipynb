<b>Answer: </b>

<b>1.</b>
This paper is about a new framework to tidy messy datasets. The goal of this research is to make data cleaning as easy and effective as possible. 

<b>2.</b>
The tidy data standard is intended to accomplish several things, such as facilitating the exploration and analysis of data, reduce data munging, and bridging the gap between the organization and analysis of data. By having the tidy data standard, analysts can keep datasets consistent, and collaboration can be streamlined between people. 


<b>3.</b>
"Like families, tidy datasets are all alike but every messy dataset is messy in its own way". This sentence is basically saying that just as families have similar characteristics that make them happy, datasets have things that make them "tidy" as well. In contrast, just as every unhappy family has problems, every messy dataset has some problems as well. 

"For a given dataset, it’s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general". This sentence is basically saying that for sp0ecific datasets, like those for houses, it is easy to figure out the observations and the attributes/features that describe those observations. However, it is very difficult to find a universal definition for observations and features because different datasets from different domains have different kinds of observations and features. 

<b>4.</b>

Values:
Values are defined as something that belongs to a variable and an observation. More specifically, the values are either numbers if they are quantitative and strings if they are qualitative.

Variables:
These are the things that contain the values. They contain the values that define a trait (such as height, temperature, and duration). As a result, they can hold both quantitative and qualitative data. 

Observations:
Observations are also things that values belong to. The observation contains all values measured on the same unit across different attributes. 


<b>5.</b>

Tidy data is defined as "a standard way of mapping the meaning of a dataset to its structure" according to section 2.3. In tidy data, each variable forms a column, each observation forms a row, each type of observational unit forms a table.

<b>6.</b>
The 5 most common problems with messy datasets are as follows: Column headers are values, not variable names. Multiple variables are stored in one column. Variables are stored in both rows and columns. Multiple types of observational units are stored in the same table. A single observational unit is stored in multiple tables.

The data in table 4 is messy because the column headers are values and not variable names. "Melting" a dataset is to turn the columns of the dataset into rows. This can help clean the data because it can turn the column headers that are values into rows instead.


<b>7.</b>

The reason that table 11 is messy is because there are multiple types in one table. For example, the fact that there is a column for every day in the month makes it messy along with the multiple types (tmax and tmin) in one column. Table 12 is tidy and molten because it consolidates the dates and values into two columns (date and value) for the molten data. The tidy data separates the multiple types in one table and makes two columns from that (tmax and tmin), eliminating the need of an element/value column as well.

<b>8.</b>
The chicken and egg problem is the interdependence between tidy data and the tools that work with it. If tidy data is only as useful as the tools that work with it, then the tools will be too specialized to the structure of the tidy data. This leads to a situation where changing the data structures or tools won't improve the workflow. This results in a state where it seems like the current methodology is optimal but is actually suboptimal.

```python
##NOTE: PART 1 HERE
import pandas as pd

air_data = pd.read_csv('data/airbnb_hw.csv')
# air_data.describe()
# air_data.head()
# print(air_data['Price'].astype(str))
print('''
There seems to be a problem with the Price variable as it is being stored as a string. If it weren't a string, then the .describe would have gotten it.
''')

#changing all the prices to numeric
air_data['Price'] = pd.to_numeric(air_data['Price'].str.replace(',', ''), errors='coerce')
# print(air_data)
print(air_data.describe())

print("""
      The price variable has been cleaned to not include commas. 
      It has also been changed so that it is of a numeric type instead of a string.
      The results of these changes are beneficial as I can also use statistics on the Prices variable now as well.""")




```

```python
# NOTE: PART 2 HERE
sharks_data = pd.read_csv('data/sharks.csv')

type_counts = sharks_data["Type"].value_counts()

print(type_counts)
print("""
    From what I can see in the values for the type column, there seem to several related types that I plan on combining into one type. 
    For example, The Boat, Boating, and Boatomg types all seem to be related so I am going to convert them into the Boating type.
    The Invalid, Questionable, Unconfirmed, Unverified, and Under investigation columns also seem to be similar so I will combine them into the Uncertain type.
    The other types seem different enough so I will keep them as is. 
    The final result of the cleaned data can be seen below.
""")
sharks_data["Type"].replace(["Boat", "Boatomg"], "Boating", inplace=True)

uncertain_types = ["Invalid", "Questionable", "Unconfirmed", "Unverified", "Under investigation"]
sharks_data["Type"].replace(uncertain_types, "Uncertain", inplace=True)

print(sharks_data)
new_type_counts = sharks_data["Type"].value_counts()
print(new_type_counts)
```

```python
#NOTE: PARTS 1 AND 2 HERE

college_data = pd.read_csv('data/college_completion.csv')

dimensions_for_college_data = college_data.shape
observations_for_college_data = dimensions_for_college_data[0]
variables_for_college_data = college_data.columns.tolist()


print(f"The dimensions_for_college_data are {dimensions_for_college_data}")
print(f"The observations_for_college_data are {observations_for_college_data}")
print(f"The variables_for_college_data are {variables_for_college_data}")
college_data.head()

```

```python
# NOTE: PART 3 HERE

cross_tabulated_data = pd.crosstab(college_data["control"], college_data["level"])
print(cross_tabulated_data)
print("""
    From this data, we can see the following patterns. Firstly, note that there are 465 2 year institutions and 527 4 year institutions 
      in the private for profit institutions. Secondly, in hte not for profit institutions, there are 68 2 year institutions but a much larger 
      number of 4 year institutions (1180). Finally, for public institutions, there are 926 2 year public institutions ands 632 4 year public 
      institutions. We can see that across the control types, there is a large difference between the number of 2 year and 4 year institutions.
      Note that the public institutions have a higher number of 2 year colleges compared to 4 year colleges.
""")

```

```python
# NOTE: PART 4 HERE:

import matplotlib.pyplot as plt
import seaborn as sns

#HISTOGRAM PLOT HEWRE
plt.figure(figsize=(12, 6))   
sns.histplot(college_data['grad_100_value'].dropna(), bins=50, color='lightgreen', label="Histogram")
plt.title('Histogram Plot of Grad 100 Values') 
plt.xlabel('Grad 100 Value')
plt.ylabel('Frequency') 
plt.show()  


# KERNEL PLOT HERE
plt.figure(figsize=(12, 6)) 
sns.kdeplot(college_data['grad_100_value'].dropna(), fill=True, color='lightblue')  
plt.title('Kernel Density Plot of Grad 100 Values')
plt.xlabel('Grad 100 Value') 
plt.ylabel('Density')  
plt.show() 

# BOXPLOT HERE
plt.figure(figsize=(12, 6))  
sns.boxplot(x=college_data['grad_100_value'], color="pink") 
plt.title('Boxplot of Grad 100 Values', ) 
plt.xlabel('Grad 100 Value')   
plt.show() 

```

```python
# PART 5 HERE:
#source: https://seaborn.pydata.org/generated/seaborn.kdeplot.html


# CONTROL GRAPHS 
plt.figure(figsize=(14, 14)) 
for control_type in college_data['control'].unique(): 
    subset = college_data[college_data['control'] == control_type]
    sns.kdeplot(subset['grad_100_value'].dropna(), fill=True, label=control_type) 
plt.title('Kernel Density Plot of Grad 100 Value by Control') 
plt.xlabel('Grad 100 Value')   
plt.ylabel('Density') 
plt.legend()  
plt.show() 

#LEVEL GRAPHS
plt.figure(figsize=(14, 14))
for level_type in college_data['level'].unique():
    subset = college_data[college_data['level'] == level_type]
    sns.kdeplot(subset['grad_100_value'].dropna(), fill=True, label=level_type)
plt.title('Kernel Density Plot of Grad 100 Value by Level')
plt.ylabel('Density')
plt.xlabel('Grad 100 Value')  
plt.legend()
plt.show()

level_description = college_data.groupby('level')["grad_100_value"].describe()
control_description = college_data.groupby('control')["grad_100_value"].describe()

print("This is the level description\n")
print(level_description)
print("\n")
print("This is the control description\n")
print(control_description)

print("From these graphs, I can see that the private not for profits tend to have higher graduation rates. Furthermore, the 4 year colleges seem to have a higher graduation rate than the 2 year colleges.")
print("""
    The institutions that appear to have the best graduation ratres are the private not-for-profit institutions, with an average graduation rate
      of 41.66 percent
""")
```

```python
# NOTE: PART 6 HERE

college_data['levelXcontrol'] = college_data['level'] + ', ' + college_data['control']
plt.figure(figsize=(14, 7))
for combination in college_data['levelXcontrol'].unique():  
    subset = college_data[college_data['levelXcontrol'] == combination]
    sns.kdeplot(subset['grad_100_value'].dropna(), fill=True, label=combination) 
plt.title('Kernel Density Plot of Grad 100 Value by levelXcontrol') 
plt.xlabel('grad_100_value') 
plt.ylabel('Density')
plt.legend()  
plt.show()
 
level_X_control_description = college_data.groupby('levelXcontrol')['grad_100_value'].describe()
print(level_X_control_description)
print('\n')
print("""From what I can see along with the .describe data, the 4 year, Private not-for-profit colleges seem to have the 
highest graduation rate with an average graduation rate of 41.91%""")
```

```python
#NOTE: PART 7 HERE:

#AID VALUE PLOTS HERE 
plt.figure(figsize=(14, 7)) 
sns.kdeplot(college_data['aid_value'].dropna(), fill=True, color='pink') 
plt.title('Kernel Density Plot of aid_value')
plt.xlabel('aid_value')
plt.ylabel('Density') 
plt.show() 


# GROUPED PLOTS HERE
plt.figure(figsize=(14, 7))
for combination in college_data['levelXcontrol'].unique():
    subset = college_data[college_data['levelXcontrol'] == combination]
    sns.kdeplot(subset['aid_value'].dropna(), fill=True, label=combination)
plt.title('Kernel Density Plot of Aid Value by Level and Control')
plt.xlabel('aid_value')
plt.ylabel('Density')
plt.legend()
plt.show() 


aid_level_descriptions = college_data.groupby('level')['aid_value'].describe()
aid_control_descriptions = college_data.groupby('control')['aid_value'].describe()

print("This is the aid level description\n")
print(aid_level_descriptions)
print("\n")
print("This is the aid control description\n")
print(aid_control_descriptions)
print("\n")
print(""" The bi-modal nature of the 'aid_value' distribution arises mainly from the differences 
      in the aid values for the 2 year and the 4 year institutions. Specifically, the local maxima created at around aid value 15000 is
      generated from the 4 year, private not-for-profit institutions which generally offers higher aid values (14180 dollars from the describe).""")
```

```python
#NOTE: PART 8 HERE:


#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
# SCATTERPLOT OF GRAD100 VALLUE BY AID VALUE HERE
plt.figure(figsize=(14, 14))
sns.scatterplot(x=college_data['aid_value'], y=college_data['grad_100_value'], alpha=0.5)  
plt.title('Scatterplot of Grad 100 Value by Aid Value')
plt.xlabel('Aid Value')   
plt.ylabel('Grad 100 Value')  
plt.show()

print("""
    For the scatterplot of Grad 100 Value by Aid Value, there seems to be a general positive trend, which might indicate that higher aid values 
    correlate with higher graduation rates. However, there is a large spread within the data, meaning that there may be a lot of variance to
    take account for.
""")

plt.figure(figsize=(14, 14))
sns.scatterplot(x=college_data['aid_value'], y=college_data['grad_100_value'], hue=college_data['level'], alpha=0.5)
plt.title('Scatterplot of Grad 100 Value by aid_value (Grouped by Level)')
plt.xlabel('aid_value')
plt.ylabel('grad_100_value')
plt.legend(title='Level')
plt.show()

print("""
    Here, we can see and grasp much more information. For scatterplots grouped by level, we can see that the 2 year institutions seem to cluster 
    in the lower range of both aid values and graduation rates. In contrast, the 4 year institutions span more aid values and graduation rates.
    We can see a stronger positive correlation between aid values and graduation rates for 4 year institutions.
""")

plt.figure(figsize=(14, 14))
sns.scatterplot(x=college_data['aid_value'], y=college_data['grad_100_value'], hue=college_data['control'], alpha=0.5)
plt.title('Scatterplot of Grad 100 Value by aid_value (Grouped by Control)')
plt.xlabel('aid_value')
plt.ylabel('grad_100_value')
plt.legend(title='Control')
plt.show()

print("""
    Once again, we can see and grasp much more information. For scatterplots grouped by control, the private for-profit institutions
    institutions primarily cluster in the lower to mid-range of aid values. Their graduation rates are more spread out but tend to be 
    lower on average. The private not-for-profit institutions show a clearer positive correlation between aid and graduation rates. 
    Many of these institutions with higher aid values also exhibit higher graduation rates. Finally, for public insitutions,
    they are mostly concentrated in the lower to mid-range of aid values, with graduation rates primarily in the lower half.
""")

print("OVERALL CONCLUSIONS:")
print("""
    From the observations that I have listed above along with the graphs that have been made, it clearly seems that 4 year, Private not-for-profit
    institutions seem to benefit the most from higher aid values in terms of graduation rates. This is consistent with the Kernel Density Plots
    and other graphs that were created above.
""")
```
